```python
import argparse
import numpy as np
from scipy.io import wavfile
import resampy
```

```python
import librosa.display
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram
%matplotlib inline
```

```python
wav_file = "C:\\Users\\Myrna\\Documents\\SensingClues\\devicehive-audio-analysis-master\\wav_files\\elephant3.wav"
data, sr = librosa.load(wav_file)
```

```python
sample_rate = 16000
```

```python
data = resampy.resample(data, sr, sample_rate)
```

```python
fig = plt.figure(figsize=(25,60), dpi = 900)
```

```python
librosa.display.waveplot(np.array(data),sr=sr)
plt.show()
```

```python
specgram(data, Fs=sr)
plt.show()
```
