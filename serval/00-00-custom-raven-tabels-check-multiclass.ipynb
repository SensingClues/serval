# In this notebook we analyse the raven tabels for the Amsterdam project 

We correct labels labeled closer then 5 seconds, so they are multiclass.

```python
# read all selection.txt files
import glob             

# location raven tabels
path = '../data/wav_samples_custom/raven-original-tags/'

txt_files = glob.glob(path + "/*selections.txt")

print(txt_files)
```

```python
# dataframe naam uit filenaam peuteren
#txt_files[0].split('/')[-1].split('.')[0]
txt_names = list(map(lambda x: x.split('/')[-1].split('.')[0], txt_files))
print(txt_names)
```

```python
# read per file a dataframe
import pandas as pd
import copy

dict_of_df = {}

for f in txt_files:

    key_name = f.split('/')[-1].split('.')[0]
    df = pd.read_csv(f, header = 0, sep='\t')
    # filter double lines some have annotated in spectral view and waveform view we only need one
    df = df.groupby('Begin Time (s)').filter(lambda x: len(x) > 1).drop_duplicates(subset='Begin Time (s)')
    # delta start
    df['dstartprevious'] = -1*df['Begin Time (s)'].diff()
    df['dstartnext'] = -1*df['Begin Time (s)'].diff(-1)

    dict_of_df[key_name] = copy.deepcopy(df)

keys = dict_of_df.keys()
```

```python
#df = dict_of_df['AS2019110741']
#dict_of_df
randomkey = list(keys)[5] # dicts are unordered
df = dict_of_df[randomkey]
```
